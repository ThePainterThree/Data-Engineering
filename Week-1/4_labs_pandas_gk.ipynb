```python
import pandas as pd

# Data Understanding
# Load the dataset into a DataFrame called df. 
df = pd.read_csv("/Users/Bootcamp/Downloads/nba.csv")
```

```python
#Display the first 5 rows. 
df.head()
```

```python
#Check the number of rows and columns. 
df.shape
```

```python
# List all column names. 
df.columns
```

```python
#Data Cleaning 
#Convert the Salary column to numeric (handle errors).
df["Salary"]= df["Salary"].astype(float)
df["Salary"].dtype
```

```python
#Count how many missing (NaN) values are in each column.
df.isna().sum()
```

```python
#Drop rows where Salary is missing.
df1 = df.dropna(subset=["Salary"])
#check
df1.isna().sum()
```

```python
#Fill missing College values with "Unknown". 
df2 = df1.copy()
df2["College"] = df1["College"].fillna("Unknown")
#check:
df2["College"].isna().sum()
```

```python
#Filtering Data
# Show all players from Boston Celtics. 
df2[df2["Team"] == "Boston Celtics"]
```

```python
#Show players whose Age is greater than 30. 
df2[df2["Age"] > 30]
```

```python
# •	Filter players who play as PG (Point Guard).
df2[df2["Position"] == "PG"]
```

```python
# •	Find players with Salary greater than 5,000,000.
df2[df2["Salary"] > 5000000]
```

```python
# Sorting
#Sort players by Salary (highest first). 
df2.sort_values("Salary", ascending= False)
```

```python
#Sort players by Age (youngest first). 
df2.sort_values("Age")
```

```python
# GroupBy Operations 
# Find the average salary per team.
df2.groupby("Team")["Salary"].mean()
```

```python
# Count number of players in each team
df2["Team"].value_counts()
```

```python
# Find the maximum salary in each team. 
df2.groupby("Team")["Salary"].max()
```

```python
# Aggregations
#What is the average age of players?
df2["Age"].mean()
```

```python
# What is the maximum salary in the dataset? 
df2["Salary"].max()
```

```python
#What is the minimum weight?
df2["Weight"].min()
```

```python
# Column Operations
# Create a new column Age_in_5_years. 
df2["Age_in_5_years"] = df2["Age"] + 5
df2.head()
```

```python
# Create a new column Salary_in_Millions (Salary / 100000). 
df2["Salary_in_Millions"] = df2["Salary" ] / 1000000
df2.head()
```

```python
# Bonus 
# Which team has the highest total salary payout? 
df2.groupby("Team")["Salary"].sum().sort_values(ascending=False)
```

```python
#Which position has the highest average salary? 
df2.groupby("Position")["Salary"].mean().sort_values(ascending=False)
```
