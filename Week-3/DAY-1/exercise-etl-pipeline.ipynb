```python
import pandas as pd

#extract data from nba.csv on local disk 
# and create an ETL pipeline
# and ELT pipeline (to do)

# 1 - EXTREACT FROM FILE
raw = pd.read_csv("./nba.csv")
raw
```

```python
# 2-TRANSFORM: ETL clean and reshape in pandas, BEFORE loading

#make a copy to keep original untouched
df = raw.copy()

#check columns names -> transform them to lowercase
df.columns = df.columns.str.strip().str.lower()
print("Columns:", df.columns.tolist())

#remove rows that were completely empty in the source data (order of execution is important)
df = df.dropna(how="all").reset_index(drop=True)

#clean column name -> missing values + reinformce type + remove trails + toLowerCase
df["name"] = (df["name"].fillna("unknown").astype(str).str.strip().str.lower())

#clean column age -> force numbers to type Int64 where NA is accepted
df["age"] = pd.to_numeric(df["age"], errors="coerce").astype("Int64")

#check if there are duplicates
print(df.duplicated().sum())

#if there were, to remove duplicates across all columns: 
# df = df.drop_duplicates()

print("Rows after cleaning:", len(df))
```

```python
#TRansform 2: validating the cleaned data
# print(df.info())
# print(df.isna().sum())

print(df.duplicated().sum())
print(df.head())
```

```python
#LOAD
df.to_csv("./nba_clean.csv", index=False)

#confirm 
loaded_df = pd.read_csv("./nba_clean.csv")

print("Cleaned rows:", len(df))
print("Loaded rows:", len(loaded_df))
print("Load successful:", len(df) == len(loaded_df))
```
